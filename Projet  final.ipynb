```python
%matplotlib inline     
import numpy as np
import random
from random import randrange
import matplotlib.pyplot as plt
from matplotlib import animation
import pandas as pd
import matplotlib
import matplotlib.animation as animation
import  matplotlib.pyplot
from random import randint
import math
import numpy

Nombre_de_bonne_poubelles = 0
Nombre_de_mauvaise_poubelles = 0

dictionaire_dechet_repartition = { "plastique" : 0.28,
     "verre" : 0.11,
     "papier" : 0.18,
     "metal" : 0.11,
     "dechets organiques " : 0.19,
     "batteries" : 0.07,
     "ampoule" : 0.05 }

dictionaire_nombre_dechet_bonne_poubelles = { "plastique" : 0,
     "verre" : 0,
     "papier" : 0,
     "metal" : 0,
     "dechets organiques " : 0,
     "batteries" : 0,
     "ampoule" : 0 }

dictionaire_nombre_dechet_mauvaise_poubelles = { "plastique" : 0,
     "verre" : 0,
     "papier" : 0,
     "metal" : 0,
     "dechets organiques " : 0,
     "batteries" : 0,
     "ampoule" : 0 }

dictionaire_taille = { "plastique" : 0,
     "verre" : 0,
     "papier" : 0,
     "metal" : 0,
     "dechets organiques " : 0,
     "batteries" : 0,
     "ampoule" : 0 }

def def_taille_poubelle(df):
    for d in df:
        var = random.randint(100,400)
        df[d]=var
    return df

def_taille_poubelle(dictionaire_taille)

def dechet_choisis(d):
    r = random.random()
    i = 1
    s = 0.0
    for k in d.keys():
        if(i==7):
            return (k)
        s += d[k]
        i +=1
        if r <= s:
            return k
        
        
def initialise_dictionnaire_avec_dechet(d,de,df,dtaille):
    poubelle = dict()
    a = False
    taille = 0
    nb_dedans = 0
    for k in dtaille:
        if(k==de):
            taille=dtaille[k]
    for k in df:
        if(k==de):
            nb_dedans =df[k]
    if(nb_dedans>(0.9*taille)):
       a = True
    for k in d.keys():
        if (k==de and a == False):
            poubelle[k] = 0.7
        if (k==de and a == True):
           poubelle[k] = 0.2
        else: 
            poubelle[k] = 0.05
    return poubelle
       

def poubelle_choisie_1(nom,d1,d2):
    v = random.random()
    if(v<=0.7):
        for k in d1:
            if(k==nom):
                d1[k]+=1
        global Nombre_de_bonne_poubelles
        Nombre_de_bonne_poubelles+=1
        return "Bonne Poubelle"
    else:
        for k in d2:
            if(k==nom):
                d2[k]+=1
        global Nombre_de_mauvaise_poubelles
        Nombre_de_mauvaise_poubelles+=1
        return "Mauvaise Poubelle"

dictionaire_nb_dechet = { "plastique" : 0,
     "verre" : 0,
     "papier" : 0,
     "metal" : 0,
     "dechets organiques " : 0,
     "batteries" : 0,
     "ampoule" : 0 }

def eboueur(d1):
    pourc_enleve = random.uniform(0.4,0.6)
    for k in d1:
        d1[k]=d1[k]-math.floor((pourc_enleve*d1[k]))
                
def un_flux(d1,d2,d3,d4,d5):
    dechet = dechet_choisis(d1)
    proba_dic = initialise_dictionnaire_avec_dechet(d1,dechet,d2,d5)
    if(poubelle_choisie_1(dechet,dictionaire_nombre_dechet_bonne_poubelles,dictionaire_nombre_dechet_mauvaise_poubelles)=="Bonne Poubelle"):
        d2[dechet]+=1
    else:
        i = 0
        for d in d2:
            if (d==dechet):
                break
            else:
                i+=1
        p = i
        while(p==i):
            p = randint(0,6)
        k = 0
        for d in d2:
            if(k==p):
                a = d
                break
            else:
                k+=1
        d2[a]+=1
        
def n_flux(n,d1):
    p = randint(40,60)
    for d in range(1,n+1):
        if (d==p):
            eboueur(d1)
        un_flux(dictionaire_dechet_repartition,dictionaire_nb_dechet,dictionaire_nombre_dechet_bonne_poubelles,dictionaire_nombre_dechet_mauvaise_poubelles,dictionaire_taille)
    return dictionaire_nb_dechet

def graphic_plot(dic):
    nb=[]
    nom=[]
    for d in dic:
        nb.append(dic[d])
    for d in dic:
        nom.append(d)
    plt.figure()
    plt.plot(nb,nom)
    plt.xlabel("Nombre de d√©chets dans la poubelle")
    plt.ylabel("Nom de la poubelle")
    plt.show()
    return 

def graphic_plot_dic_bonne(dic):
    nb=[]
    nom=[]
    for d in dic:
        nb.append(dic[d])
    for d in dic:
        nom.append(d)
    plt.figure()
    plt.plot(nb,nom)
    plt.xlabel("Nombre de bon d√©chets dans la poubelle")
    plt.ylabel("Nom de la poubelle")
    plt.show()
    return   

def graphic_plot_dic_mauvaise(dic):
    nb=[]
    nom=[]
    for d in dic:
        nb.append(dic[d])
    for d in dic:
        nom.append(d)
    plt.figure()
    plt.plot(nb,nom)
    plt.xlabel("Nombre de mauvais d√©chets dans la poubelle")
    plt.ylabel("Nom de la poubelle")
    plt.show()
    return 

def batonnets_plot(dic):
    nb=[]
    nom=[]
    for d in dic:
        nb.append(dic[d])
    for d in dic:
        nom.append(d)
    plt.figure()
    plt.bar(nom,nb)
    plt.xticks(rotation=40)
    plt.xlabel("Type de la poubelle")
    plt.ylabel("Nombre de d√©chets dans la poubelle")
    plt.show()
    return

def camenbert(dic):
    nom = []
    nb = []
    for d in dic:
        nb.append(dic[d])
    for d in dic:
        nom.append(d)
    x = nb
    plt.figure(figsize = (8, 8))
    plt.pie(x, labels = nom,
           autopct = lambda x : str(round(x, 2)) + '%',
           pctdistance = 0.7, labeldistance = 1.1)

    plt.legend(bbox_to_anchor=(1, 1),
           bbox_transform=plt.gcf().transFigure)
    return 

             
def test_dict_programme(d1,d2):
    s = 0.0
    for d in d1:
        s += d1[d]
    for d in d2:
        s += d2[d]
    global Nombre_de_bonne_poubelles
    global Nombre_de_mauvaise_poublles
    if(s==(Nombre_de_bonne_poubelles+Nombre_de_mauvaise_poubelles)):
        return "Tout va bien, tous les tests sont bon ! :)"
    else:
        return "Error test_dict_programme"

def main(n):
    n_flux(n,dictionaire_nb_dechet)
    if(test_dict_programme(dictionaire_nombre_dechet_bonne_poubelles,dictionaire_nombre_dechet_mauvaise_poubelles)=="Tout va bien, tous les tests sont bon ! :)"):
        a = True
    else:
        a = False
    if(a==True):
        print("Graphique pour le nb de d√©chet :")
        graphic_plot(dictionaire_nb_dechet)
        print("Graphique pour le nb de bon d√©chets dans les poubelles :")
        graphic_plot_dic_bonne(dictionaire_nombre_dechet_bonne_poubelles)
        print("Graphique pour le nb de mauvais d√©chet dans les poubelles :")
        graphic_plot_dic_mauvaise(dictionaire_nombre_dechet_mauvaise_poubelles)
        print("Graphique pour le nb de d√©chet avec les batonnets :")
        batonnets_plot(dictionaire_nb_dechet)
        print("Graphique pour le nb de bon d√©chet avec les batonnets :")
        batonnets_plot(dictionaire_nombre_dechet_bonne_poubelles)
        print("Graphique pour le nb de mauvais d√©chet avec les batoonets :")
        batonnets_plot(dictionaire_nombre_dechet_mauvaise_poubelles)
        print("1- Camenbert pour le nb de d√©chet en % :")
        camenbert(dictionaire_nb_dechet)
        print("2- Camenbert pour le nb de bon d√©chet en % :")
        camenbert(dictionaire_nombre_dechet_bonne_poubelles)
        print("3- Camenbert pour le nb de mauvais d√©chets en % :")
        camenbert(dictionaire_nombre_dechet_mauvaise_poubelles)
        
        
main(5160)


    
```

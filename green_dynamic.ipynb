```python
%matplotlib inline     
# ^Sans animations
# %matplotlib notebook 
# ^Avec animations
import numpy as np
import random
from random import randrange
import matplotlib.pyplot as plt
from matplotlib import animation
import pandas as pd
import matplotlib
import matplotlib.animation as animation
import  matplotlib.pyplot
```

```python
d = { "Plastique" : 0.28,
     "Verre" : 0.11,
     "Papier" : 0.18,
     "Metal" : 0.11,
     "Organiques " : 0.19,
     "Batteries" : 0.07,
     "Ampoule" : 0.05 }  

def dechet(d):
    r = (randrange(0,100)*1.0)/100
    s = 0
    for k in d.keys():
        s += d[k]
        if r < s:
            return k
        
dechet(d)
```

```python
def pou(d, de):
    poubelle = dict()
    for k in d.keys():
        if k == de:
            poubelle[k] = 0.7
        else: 
            poubelle[k] = 0.05
    return poubelle
       

x = dechet(d)
pou(d, x)
```

```python
def poubelle_choisie(d):
    de = dechet(d)
    r = (randrange(0,100)*1.0)/100
    s = 0
    dic = pou(d, de)
    for k in dic.keys():
        s+= dic[k]
        if r < s:
            if k == de :
                return "Bonne Poubelle : " + de
            else:
                return "Mauvaise Poubelle : " + k
                
poubelle_choisie(d)
```

```python
def nb_dechet_poubelle(d, n):
    nb = dict()
    for x in d.keys():
        nb[x] = 0
        
    i = 0
    while ( i < n ):
        a = poubelle_choisie(d)
        for x in d.keys():
            if a == "Bonne Poubelle : " + x or a == "Mauvaise Poubelle : " + x:
                nb[x] += 1

        i = i + 1
        
    return nb
    
nb_dechet_poubelle(d, 4000)
    
```

```python
def proportion(d, n):
    i = 0
    prop = {"Bonne Poubelle" : 0,
           "Mauvaise Poubelle" : 0}
    while ( i <= n ): 
        a = poubelle_choisie(d)
        for x in d.keys():
            if a == "Bonne Poubelle : " + x:
                prop["Bonne Poubelle"] += 1
            if a == "Mauvaise Poubelle : " + x:
                prop["Mauvaise Poubelle"] += 1
        i = i + 1
        
        
    return prop

proportion(d, 400)
```

```python
def camenbert(dic):
    nom = []
    nb = []
    for d in dic:
        nb.append(dic[d])
    for d in dic:
        nom.append(d)
    x = nb
    plt.figure(figsize = (8, 8))
    plt.pie(x, labels = nom,
           autopct = lambda x : str(round(x, 2)) + '%',
           pctdistance = 0.7, labeldistance = 1.1)

    plt.legend(bbox_to_anchor=(1, 1),
           bbox_transform=plt.gcf().transFigure)
    return 


lol = nb_dechet_poubelle(d, 4000)
print(lol)
camenbert(lol)
```

```python
def graphic_plot(dic):
    nb=[]
    nom=[]
    for d in dic:
        nb.append(dic[d])
    for d in dic:
        nom.append(d)
    plt.figure()
    plt.bar(nom,nb)
    plt.xticks(rotation=40)
    plt.xlabel("Type de la poubelle")
    plt.ylabel("Nombre de déchets dans la poubelle")
    plt.show()
    return

graphic_plot(nb_dechet_poubelle(d,4000))
    
```
